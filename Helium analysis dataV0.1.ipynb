```python
import os
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits
```

```python
A = 6.6626452*10**-10
B = 2.7302378*10**-7
```

```python
image_data=1
tracks = []
n = 0
for i in os.listdir('/Users/Edward/Documents/year 2/lab project/Ed-Em-Y2Spec/Helium data/'):
    if i.endswith(".fit"):
        n += 1
        
        filename = '/Users/Edward/Documents/year 2/lab project/Ed-Em-Y2Spec/Helium data/'+ str(i)

        image_file = get_pkg_data_filename(filename)

        image_data = fits.getdata(image_file, ext=0)
        tracks.append(image_data)

#print(tracks)        
mean = np.mean(tracks, axis = 0) #gets the mean of every single list, compiling it into a mean picture
snipmean = mean[210:410]

x = np.linspace(1,800, 800)
x = A*x + B
y = np.mean(snipmean, axis = 0)
wavelengthlist = [388.8648*10**-9, 402.618*10**-9, 447.148*10**-9, 501.57*10**-9, 587.5615*10**-9, 656.01*10**-9, 667.85*10**-9, 706.5*10**-9]
for k in wavelengthlist:
    plt.axvline(x=k,color='r', linestyle='--')
#plt.axvline(x=388*10**-9,color='r', linestyle='--')
#plt.axvline(x=501*10**-9,color='r', linestyle='--')
#plt.axvline(x=587*10**-9,color='r', linestyle='--')
#plt.axvline(x=668*10**-9,color='r', linestyle='--')
#plt.axvline(x=706*10**-9,color='r', linestyle='--')
plt.title('Plot of Mean Intensity Vs Wavelength')
plt.xlabel(r'Wavelength (m)')
plt.ylabel(r'Intensity(arbitrary units)')
plt.plot(x, y)
plt.show()
print(n)
```
